```
# Sales Îç∞Ïù¥ÌÑ∞ÌÉÄÏûÖÏùÑ floatÎ°ú Î≥ÄÍ≤Ω, Îã®ÏúÑÎäî K Î°ú ÌÜµÏùº
def toInt(x):
  if x.endswith('K'):
    x = x.replace('K','')
    x = pd.to_numeric(x)
    return x

  elif x.endswith('M'):
    x = x.replace('M','')
    x = pd.to_numeric(x)
    x = 1000 * x
    return x

  else :
    x = pd.to_numeric(x)
    x = 1000 * x
    return x
```

```
df['NA_Sales'] = df['NA_Sales'].apply(toInt)
df['EU_Sales'] = df['EU_Sales'].apply(toInt)
df['JP_Sales'] = df['JP_Sales'].apply(toInt)
df['Other_Sales'] = df['Other_Sales'].apply(toInt)
```

```
# ÌòÑÏû¨ÍπåÏßÄ Ï∂úÏãúÍ∏∞Í∞Ñ Íµ¨ÌïòÍ∏∞ ÏúÑÌïú Ìï®Ïàò
def period(x):
  p = 2020 - x + 1 
  return p
```

```
df_high['Period'] = df_high['Year'].apply(period)
```

```
df_high['Sales_per_period'] = df_high['Total_Sales']/df_high['Period']
```

```
df_high.head()
```

```
spp = df_high.sort_values('Sales_per_period').nlargest(61,'Sales_per_period')
spp.head()
```

```
spp_over_2010 = spp[spp['Year'] >= 2011]
spp_over_2010
```

```
fig = px.scatter(spp_over_2010, x="Year", y="Total_Sales", color="Genre",
                 size='Total_Sales',hover_data =['Name'])

fig.update_layout(title="(Over $10,000,000)Ïó∞ÎèÑÎ≥Ñ Îß§Ï∂úÏï° Î∂ÑÌè¨",titlefont_size=20, xaxis_title="Years",
    yaxis_title="High Sales")

fig.show()
```
